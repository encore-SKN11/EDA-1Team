5. 데이터 정제 및 전처리
    - 필요 없는 변수나 중복 데이터를 제거한다.
    - 범주형 데이터를 처리하거나, 스케일링 및 정규화를 통해 모델에 적합한 형태로 데이터를 변환한다. 
    - df.drop(), df.fillna(), pd.get_dummies() 등의 함수를 활용한다.

```python
df_new = df.drop(columns=['이름', '득점', '2루타', '3루타', '타점', '도루', '볼넷', '사구', '고의사구', '삼진', '병살', '희생타', '희생플라이'])
df_avg = pd.read_csv('yearly_averages.csv')
# display(df_avg)
# display(df_new)

merged = pd.merge(df_new, df_avg, on="연도", suffixes=('_all', '_avg'))

# display(merged)

columns_to_subtract = ['WAR', '타율', '홈런', '장타', 'OPS', 'wRC+', '안타', '출루']

for col in columns_to_subtract:
    merged[f'{col}'] = merged[f'{col}_all'] - merged[f'{col}_avg']

# 차이값만 모은 DataFrame 생성
df_diff = merged[['연도'] + [f'{col}' for col in columns_to_subtract]+['수상여부']]



df_diff = df_diff.drop(columns=['연도'])
# 결과 확인
# display(df_diff)

X = df_diff.drop(columns=['수상여부'])

y = df_diff['수상여부']

# 정규화
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# display(X_scaled, y)
# df_new[df_new['수상여부']==1].value_counts
print(df['수상여부'].value_counts())

```

## 학습 구간(knn)

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

# 학습/테스트 분리
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.25, random_state=55)

# 모델 정의 및 학습
model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)

# 예측
y_pred = model.predict(X_test)

# 결과 평가
print("정확도:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

```

```python
# 원본 데이터에서 '이름' 컬럼 포함한 데이터 만들기
df_with_names = df[['이름', '연도', '수상여부'] + columns_to_subtract]

# '연도' 기준으로 yearly_averages와 병합
merged_with_names = pd.merge(df_with_names, df_avg, on="연도", suffixes=('_all', '_avg'))

# 차이값 계산
for col in columns_to_subtract:
    merged_with_names[f'{col}'] = merged_with_names[f'{col}_all'] - merged_with_names[f'{col}_avg']

# 필요한 컬럼 선택 (이름 포함)
df_diff_with_names = merged_with_names[['이름', '연도'] + [f'{col}' for col in columns_to_subtract] + ['수상여부']]

# '연도' 제외
df_diff_with_names = df_diff_with_names.drop(columns=['연도'])

# X, y 분리
X_with_names = df_diff_with_names.drop(columns=['이름', '수상여부'])
y_with_names = df_diff_with_names['수상여부']
names = df_diff_with_names['이름']  # 선수 이름 저장

# 정규화
scaler = StandardScaler()
X_scaled_with_names = scaler.fit_transform(X_with_names)

# 학습/테스트 분리 (선수 이름도 함께 분리)
X_train, X_test, y_train, y_test, names_train, names_test = train_test_split(
    X_scaled_with_names, y_with_names, names, test_size=0.25, random_state=55
) # 3, 55

# 모델 학습
model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)

# 예측
y_pred = model.predict(X_test)

# 결과 데이터프레임 생성
results_df = pd.DataFrame({
    '이름': names_test,
    '실제 수상여부': y_test.values,
    '예측 수상여부': y_pred
})

# 결과 출력
pd.set_option('display.max_rows', None)
print(results_df)

# 모델 성능 평가
print("정확도:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
